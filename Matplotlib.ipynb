```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
```

```python
x=np.array([2022,2023,2024,2025])
y=np.array([0,0,14,25])
plt.plot(x,y,marker="v",
         markersize=5, 
         markerfacecolor="Red",
         markeredgecolor="Red",
         linestyle="solid",
         linewidth=2,
         color="Blue")
plt.show()
```

```python
x=np.array([2022,2023,2024,2025])
y1=np.array([0,0,14,25])
y2=np.array([2,4,5,6])
y3=np.array([45,7,5,4])
linestyle=dict(marker="v",
         markersize=5, 
         markerfacecolor="Red",
         markeredgecolor="Red",
         linestyle="solid",
         linewidth=2,
         color="Blue")
plt.plot(x,y1,**linestyle)
plt.plot(x,y2,**linestyle)
plt.plot(x,y3,**linestyle)
plt.show()
```

```python
x=np.array([2022,2023,2024,2025])
y=np.array([0,0,14,25])
plt.grid(linewidth=1,color="blue",
         linestyle="dashed")
plt.plot(x,y,marker="o")
plt.show()
```

```python
fruits=np.array(["Apple","Orange","Grapes","Dates"])
sales=np.array([120,345,690,324,])
line_style=dict(fontsize=15,
                family="Arial",
                fontweight="bold",
                color="blue")
plt.bar(fruits,sales,color="red")
plt.title("Fruit Sales",fontsize=25,color="red")
plt.xlabel("Fruits", **line_style)
plt.ylabel("Sales", **line_style)
plt.show()
```

```python
sales=np.array([120,345,690,324])
mylabels=["Apple","Bananas","cherries","papaya"]

plt.pie(sales,labels=mylabels,autopct="%1.1f%%")
plt.title("Fruit Sales")
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,4])
plt.xlabel("X",color="red")
plt.ylabel("Y",color="blue")
plt.title("SCATTER")
plt.scatter(x,y)
plt.show()
```

```python
x=np.random.normal(loc=70,scale=10,size=100)
plt.hist(x)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
figure,axes=plt.subplots(2,2)

axes[0,0].plot(x,x*2,color="Red")
axes[0,0].set_title("x*2")

axes[0,1].plot(x,x**2,color="Blue")
axes[0,1].set_title("x**2")

axes[1,0].plot(x,x**3,color="Yellow")
axes[1,0].set_title("x**3")

axes[1,1].plot(x,x**4,color="Green")
axes[1,1].set_title("x**4")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("orders.csv")

line_style=dict(marker="o",color="blue")
catergory=df.groupby('Category')['Quantity'].sum()
print(catergory)
```

```python
plt.plot(catergory.index,catergory.values)
plt.show()
```

```python
plt.bar(catergory.index,catergory.values)
plt.show()
```

```python
sk=np.array(["Electronics","Furniture","Stationery"])
plt.pie(catergory.values,labels=sk,autopct="%1.1f%%")
plt.show()
```
